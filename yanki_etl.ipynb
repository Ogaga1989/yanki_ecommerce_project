#### Data Cleaning and Transformation

```python
yanki_df.loc[yanki_df['State'].isna()].head()
```

```python
# Drop missing values
yanki_df.dropna(subset=['Order_ID','Customer_ID'], inplace=True)
```

```python
# Checking if missing value has been dropped

yanki_df.info()
```

```python
# Convert Order_Date from string to datatime object
yanki_df['Order_Date'] = pd.to_datetime(yanki_df['Order_Date'], format='%d/%m/%Y %H:%M')
```

```python
yanki_df.info()
```

```python
# Resetting index
yanki_df.reset_index(drop=True, inplace=True)
```

```python
# Customer Table
# We need to create a copy so that whatever happens to the dataset in the future does not affect the database
Customer_df = yanki_df[['Customer_ID', 'Customer_Name', 'Email', 'Phone_Number' ]]\
    .copy().drop_duplicates().reset_index(drop=True)

# View the Customer Table
Customer_df.head()
```

```python
# Product Table
Product_df = yanki_df[['Product_ID','Product_Name', 'Brand', 'Category', 'Price']]\
    .copy().drop_duplicates().reset_index(drop=True)


# View the Product Table
Product_df.head()
```

```python
#  Order Table
Order_df = yanki_df[['Order_ID','Customer_ID', 'Product_ID', 'Quantity', 'Total_Price', 'Order_Date']]\
    .copy().drop_duplicates().reset_index(drop=True)


# View the Order Table
Order_df.head()
```

```python
# Shipping Table
Shipping_Address_df = yanki_df[['Customer_ID','Order_ID', 'Shipping_Address', 'City', 'State','Country', 'Postal_Code']]\
    .copy().drop_duplicates().reset_index(drop=True)

# Naming the index as Shipping_ID
Shipping_Address_df.index.name = 'Shipping_ID'

# Reset the index
Shipping_Address_df.reset_index(inplace=True)

# View the Shipping Table
Shipping_Address_df.head()
```

```python
# Payment Table
Payment_df = yanki_df[['Order_ID','Payment_Method', 'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)


# View the Payment Table
Payment_df.head()
```

```python
# Save the tables to csv files
Customer_df.to_csv(r'Dataset\Cleaned_Data\customers.csv', index=False)
Product_df.to_csv(r'Dataset\Cleaned_Data\products.csv', index=False)
Order_df.to_csv(r'Dataset\Cleaned_Data\orders.csv', index=False)
Shipping_Address_df.to_csv(r'Dataset\Cleaned_Data\shipping_address.csv', index=False)
Payment_df.to_csv(r'Dataset\Cleaned_Data\payments.csv', index=False)
```
